```python
import pandas as pd
import matplotlib as plt
import numpy as np
from scipy.stats import linregress
```

```python
dogRates = pd.read_csv('dog_rates_tweets.csv', parse_dates=['created_at'])
```

```python
dogRates['rating'] = 0
dogRates['rating'] = dogRates['text'].str.extract(r'(\d+(\.\d+)?)/10').astype('double')
```

```python
dogRates = dogRates[dogRates['text'].str.contains(r'[2-9][0-5]/10', regex=True)==False]
dogRates = dogRates.dropna()
dogRates = dogRates.drop(dogRates[dogRates.rating > 20].index)
```

```python
# code gotten from jezrael from stackover flow
# https://stackoverflow.com/questions/40881876/python-pandas-convert-datetime-to-timestamp-effectively-through-dt-accessor
dogRates['timestamp'] = dogRates['created_at'].values.astype(np.double) // 10**9
fit = linregress(dogRates.timestamp, dogRates.rating)
dogRates['prediction'] = (fit.slope*dogRates['timestamp'] + fit.intercept)
```

```python
plt.pyplot.plot(dogRates['created_at'], dogRates['rating'], 'b.', alpha=0.5)
plt.pyplot.plot(dogRates.created_at, fit.slope*dogRates['timestamp'] + fit.intercept, 'r-', linewidth=3)
plt.pyplot.xticks(rotation=25)
plt.pyplot.title("User's Dog Picture Ratings")
plt.pyplot.xlabel("Date")
plt.pyplot.ylabel("Rating")
plt.pyplot.show()
```
